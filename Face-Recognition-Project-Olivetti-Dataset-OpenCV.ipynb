#**Face Recognition using OpenCV on `Olivetti Dataset`**

###**Import the necessary packages**

```python
import cv2
import matplotlib.pyplot as plt
```

###**Get a Pre-Trained Classifier for face detection**

OpenCV has a lot of pre-trained classifiers for face detection, eye detection etc...
We can get these `.xml` files from [GitHub.](https://github.com/opencv/opencv/tree/master/data/haarcascades)

It has several positive and negative samples to train the model (Viola-Jones Algorithm)

```python
cascade_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_alt.xml")
```

###**Open an Image**

```python
# Open an image and show it
#image = cv2.imread('Girl_face.jpg')
#image = cv2.imread('6 faces.png')
image = cv2.imread('15 faces.jpg')
```

As OpenCV deals with BGR and matplotlib deals with RGB we observe slight distortions in the image.

To get the image correctly we have to make the transformation.

```python
# Making the transformation
#image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Convert the image into GRAYSCALE
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

As we have to work with the GRAYSCALE images in Face Detection we can directly convert the image from BGR to GRAY without converting it into RGB.

Now using the already defined cascade classifier we have to detect the number of faces in a given image.

```python
detected_faces = cascade_classifier.detectMultiScale(
    gray_image, 
    scaleFactor=1.1, 
    minNeighbors=10, 
    minSize=(30, 30)
)
```

`detected_faces` gives the number of faces in the given image and their coordinates to plot the rectangular box around the faces.

Now we will write a `for` loop for detecting and drawing a rectangle around the faces

```python
for (x, y, width, height) in detected_faces:
  cv2.rectangle(image, (x, y), (x + width, y + height), (0, 255, 0), 5)
```

Convert the BGR image to RGB color and show the image

```python
print(f"Total {len(detected_faces)} face(s) detected.")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
